```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.salesmate.io/blog/consulting-sales-pipeline/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


title = soup.h1.text.strip()
content = soup.find('div', class_='blog-contain-box-inner').text.strip()
print(f'Title: {title}\nContent: {content}\n')
```

```python

for i in range(1, 53):
    url = f'https://www.salesmate.io/blog/page/{i}/'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    try:
        for article in soup.find_all('a', class_='blog-title'):
            link = article['href']
            response = requests.get(link)
            soup = BeautifulSoup(response.text, 'html.parser')
            title = soup.h1.text.strip()
            content = soup.find('div', class_='blog-contain-box-inner').text.strip()
            # add data to json file
            # truncate the content to 2000 words
            if len(content) > 5000:
                content = content[:5000]
                
            data['blogs'].append({ 'title': title, 'content': content })
    except:
        # save data to json file
        with open('salesmate_doc.json', 'w') as f:
            json.dump(data, f, indent=4)
 
```
